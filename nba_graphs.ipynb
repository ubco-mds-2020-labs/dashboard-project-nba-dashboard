```python
# See link on how to create a legend
# https://github.com/altair-viz/altair/issues/984

alt.layer(
    alt.Chart(df).mark_line(color = 'red', size = 1).encode(y = alt.Y('TOV',axis=alt.Axis(title='Count', grid =False))),
    alt.Chart(df).mark_line(color = 'blue', size = 1).encode(y = 'REB'),
    alt.Chart(df).mark_line(color = 'green', size = 1).encode(y = 'AST')  
).encode(x = alt.X('Season',axis=alt.Axis(title='Season'))
).configure_view(
    stroke='transparent'
).properties(
    height = 160,
    width = 384
).configure_axis(
    labelFontSize=10,
    titleFontSize=12
)
```
